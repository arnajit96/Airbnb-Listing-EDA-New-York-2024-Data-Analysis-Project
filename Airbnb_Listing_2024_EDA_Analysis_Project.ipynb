**Task 5: Data Analysis**


**Univerient Analysis**

```python
# Price Distribution
data['price']
```

```python
sns.histplot(data=data,x='price')
```

```python
#Price Distribution

sns.boxplot(data=data,x='price')
```

```python
# In upper boxplot diagram one price which is included in 100000
# so we can using some condition where the price is less than 1500
# that can be easily find the price range of data

df=data[data['price']<1500]
sns.boxplot(data=df,x='price')
```

```python
sns.histplot(data=df,x='price')
```

```python
plt.figure(figsize=(8,5)) # Change size of figure
sns.histplot(data=df,x='price',bins=100)
plt.title("Price Distribution")
plt.ylabel("Frequency") #Change the y axis name of Frequency instead of price
plt.show()
```

```python
# heat map- corrlation of one variable with others for numerical column
corr =df[['latitude','longitude','price',
       'minimum_nights','number_of_reviews','reviews_per_month',
       'availability_365','beds']].corr()

corr
plt.figure(figsize=(8,6))
sns.heatmap(data=corr,annot=True)
```
